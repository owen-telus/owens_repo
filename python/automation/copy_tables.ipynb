# Copy Tables from one wb to another

Get Table info

```python
%%bigquery df
SELECT *
FROM `cto-wln-sa-data-pr-bb5283`.`region-northamerica-northeast1`.INFORMATION_SCHEMA.TABLES
WHERE table_schema IN ('5G_speed_tiers', 'measurements_data') AND table_type = 'BASE TABLE'
ORDER BY 
table_type,
creation_time
```

```python
df
```

```python
new_wb_name = 'wb-ai-accelerator-1-pr-71edf3'
```

```python
for index,row in df.iterrows():
    query = \
    '''
        CREATE TABLE `{wb_name_to}.{dataset_name}.{table_name}` AS (
            SELECT * FROM `{wb_name_from}.{dataset_name}.{table_name}`
        );
    '''.format(wb_name_from = row['table_catalog'],
               wb_name_to =new_wb_name,
               dataset_name = row['table_schema'],
               table_name = row['table_name']
              )
    print(query)
```
